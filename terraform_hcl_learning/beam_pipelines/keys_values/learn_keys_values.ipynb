```python
import apache_beam as beam

#key-value pair pvalue in beam is present by a tuple.
#using beam.Keys() to get the key of that pvalue
with beam.Pipeline() as pipeline:
  icons = (
      pipeline
      | 'Garden plants' >> beam.Create([
          ('ðŸ“', 'Strawberry'),
          ('ðŸ¥•', 'Carrot'),
          ('ðŸ†', 'Eggplant'),
          ('ðŸ…', 'Tomato'),
          ('ðŸ¥”', 'Potato'),
      ])
      | 'Keys' >> beam.Keys()
      | beam.Map(print))
```

```python
import apache_beam as beam

#using beam.Values() transform to get value of key-value pvalue in beam
with beam.Pipeline() as pipeline:
  plants = (
      pipeline
      | 'Garden plants' >> beam.Create([
          ('ðŸ“', 'Strawberry'),
          ('ðŸ¥•', 'Carrot'),
          ('ðŸ†', 'Eggplant'),
          ('ðŸ…', 'Tomato'),
          ('ðŸ¥”', 'Potato'),
      ])
      | 'Values' >> beam.Values()
      | beam.Map(print))
```

```python
import apache_beam as beam

#using beam.KvSwap() to swap key-value of pvalue
with beam.Pipeline() as pipeline:
  plants = (
      pipeline
      | 'Garden plants' >> beam.Create([
          ('ðŸ“', 'Strawberry'),
          ('ðŸ¥•', 'Carrot'),
          ('ðŸ†', 'Eggplant'),
          ('ðŸ…', 'Tomato'),
          ('ðŸ¥”', 'Potato'),
      ])
      | 'Key-Value swap' >> beam.KvSwap()
      | beam.Map(print))
```
